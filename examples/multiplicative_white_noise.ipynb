# Example: Multiplicative White Noise Process

This one-dimensional version of the this process $X(t)$ is defined by the stochastic differential equation 

$$dX = \theta X dW$$

where $\theta$ is a constant parameter and $dW$ is the differential of the Wiener Process. This is an Ito-stochastic differential equation so the increments of the Wiener Process are independent of the state $X$ on the r.h.s. of the equation. Multiplicative noise is very common in stochastic population dynamics and chemical reactions.

The solution can be obtained by integration and reads

$$X(t)=X(t_0)\exp\left[W(t)-W(t_0)\right]-\frac{1}{2}\theta^2 (t-t_0)$$

so if we let $X(t_0)=1$, $t_0=0$ we get

$$X(t)=\exp\left[W(t)\right]-\frac{1}{2}\theta^2 t$$

This means that the expetation value of $\left<X(t)\right>=1$ so it remains constant and doesn't change. However the probability density function $p(x,t)$ changes over time. In fact, when you look at realizations of the process you see that most trajectories decrease and approach zero, and a few trajectories make long excursions.


```python
import stopro as sp
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Here are a few realizations, first on a linear scale and then on a log scale

```python
T = 10
dt = 0.01
samples = 20
theta = 1.0
ens = sp.multiplicative_white_noise(T,dt,samples=samples,theta=theta)
```

```python
fig,ax = plt.subplots(figsize=[8,4])
for i in range(len(ens["X"])):
    ax.plot(ens["t"],np.squeeze(ens["X"][i]))
    
plt.xlabel("t")
plt.ylabel("X(t)")
plt.show()
```

```python
fig,ax = plt.subplots(figsize=[8,4])
for i in range(len(ens["X"])):
    ax.plot(ens["t"],np.squeeze(ens["X"][i]))
    
plt.xlabel("t")
plt.ylabel("X(t)")
plt.yscale("log")
plt.show()
```
