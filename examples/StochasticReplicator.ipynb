# The stochastic replicator model

The stochastic replicator model is stochastic process that generalized the replicator model. The replicator model is a deterministic model for the fraction of species $n=1,...,N$ governed by 

$$\dot x_i = (\mu_i-\phi)x_i$$

where the parameters $\mu_i$ are fitnesses of species $i$ and $\phi=\sum _i \mu_i x_i$ is the population mean fitness and the population is normalized to unity, i.e. $\sum x_i=1$ because the $x_i$ are fractions of the entire population.

The stochastic replicator model is a limiting case for both, the **Kimura replicator** model and the **Gillespie replicator** model, when instead of Gaussian white noise, **colored noise** with non-zero correlation time is considered and when this correlation time tends to zero. This is equivalent to the Stratonovich interpretation of the associated stochastic differential equations.

So, specifically the model is defined as

$$\dot x_i = (\mu_i(t)-\phi)x_i$$

with 

$$\mu_i(t)=\mu_i + \sigma_i Z(t)_i $$

and

$$\tau dZ_i=-Z_i dt+dW_i$$

and one considers the limit $\tau\rightarrow0$ after integration. Using the Gillespie model, one finds the solution:

$$x_i(t)=\frac{1}{1+\sum_{j\neq i} \frac{x_j(0)}{x_i(0)} \exp\left[\left(\mu_j-\mu_i \right) t +\sigma_j W_j(t) -  \sigma_i W_i(t)\right]   }$$

because the integration of $\mu_i(t)$ is the integral of the Ornstein-Uhlenbeck process, in the limit $\tau\rightarrow0$ this is the Wiener process.

For two populations we've got an equivalence to the SDE:

$$dx = x(1-x)\left[\mu_1-\mu_2+\sigma^2(1/2-x))\right]dt +\sigma x(1-x) dW$$

so in other words, this is a Gillespie model where we replaced $\mu_i\rightarrow \mu_i+\sigma_i^2/2$.

```python
import matplotlib.pyplot as plt
import numpy as np
import stopro as p
```

## n populations with identical parameters

```python
samples = 2
n = 5
mu = np.ones(n)
sigma = np.ones(n)

res = p.stochastic_replicator(10,0.001,N=5,samples=samples,gap=10)

fig,ax = plt.subplots(samples,1,figsize=[10,6])
for i in range(samples):
    for j in range(n):
        ax[i].plot(res["t"],np.squeeze(res["X"][i][j]))
        ax[i].set_xlabel("t")
        ax[i].set_ylabel("X(t)")
plt.show()
```

## three populations with different noise terms and identical mean fitness

```python
samples = 3
n = 3
mu = np.ones(n)
sigma = np.array([0.1,1.0,2.0])

res = p.stochastic_replicator(10,0.001,N=3,sigma=sigma,samples=samples,gap=10)

leg=[r"$\mu$ ="+f'{x:.1f}'+r", $\sigma$ ="+f'{y:.1f}' for (x,y) in zip(res["mu"],res["sigma"])]

fig,ax = plt.subplots(samples,1,figsize=[10,6])
for i in range(samples):
    for j in range(n):
        ax[i].plot(res["t"],np.squeeze(res["X"][i][j]))
        ax[i].set_xlabel("t")
        ax[i].set_ylabel("X(t)")
        ax[i].legend(leg)
fig.tight_layout()
plt.show()
```

## two populations, many realizations

```python
T = 10
samples = 200
N = 2
sigma = 2

res = p.stochastic_replicator(T,0.001,N=N,sigma=sigma,samples=samples,gap=10)

fig,ax = plt.subplots(1,1,figsize=[10,4])
for i in range(samples):
    ax.plot(res["t"],np.squeeze(res["X"][i][0]),color="C0",alpha=0.1)
    ax.set_xlabel("t")
    ax.set_ylabel("X(t)")

plt.show()
```
