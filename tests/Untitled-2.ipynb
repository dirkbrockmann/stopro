For correlated driving noise with diagonal drift theta, the stationary covariance (P) solves the Lyapunov equation

$$
\Theta P + P\Theta = \Sigma,C,\Sigma,
\quad \Theta=\mathrm{diag}(\theta),\ \Sigma=\mathrm{diag}(\sigma),
$$

which gives the elementwise closed form:

$$
P_{ik}=\frac{\sigma_i \sigma_k, C_{ik}}{\theta_i+\theta_k}.
$$

So you should sample ($X(0)\sim \mathcal N(0,P)$).

Below is a minimal patch that only fixes the stationary initializer (keeps your Euler update and your dw generation unchanged).

```python
stopro.wiener(1, dt=0.3,N=3,samples=2,covariance=[[1,0.5,0.2],[0.5,1,0.3],[0.2,0.3,1]])
```
